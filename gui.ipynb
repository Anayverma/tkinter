```python
from tkinter import *
from tkinter import messagebox

def login():
    # Clear previous labels
    clear_labels()

    # Get the values from the Entry widgets
    user_name = t1.get()
    password = t2.get()

    # Validate input
    if not user_name or not password:
        messagebox.showerror("Error", "Please enter both user name and password.")
    else:
        # Perform login logic here
        Label(box, text=f"Welcome, {user_name}!").pack()

def exit_e():
    # Clear previous labels
    clear_labels()
    Label(box, text="Are you sure you want to exit? ").pack()

def clear_labels():
    # Destroy all widgets in the box (including labels)
    for widget in box.winfo_children():
        widget.destroy()

box = Tk()
box.geometry('400x400')
# box.configure(bg='red')
box.title("Login-Page")

Label(box, text="USER_NAME").grid(row=0, column=0)
Label(box, text="PASSWORD").grid(row=1, column=0)

t1 = Entry(box)
t2 = Entry(box, show="*")  # Use show="*" to hide password input
t1.grid(row=0, column=1, padx=5)
t2.grid(row=1, column=1, padx=5)

b_login = Button(box, text="Login", bg='yellow', fg='orange', command=login)
b_login.grid(row=2, column=0, pady=10)

b_exit = Button(box, text="Exit", bg='yellow', fg='orange', command=exit_e)
b_exit.grid(row=2, column=1, pady=10)

mainloop()
```

```python
# add a listbox widget having some choices of cities to be visited from those choices only two choices needed to be submitted others need to be eliminated

from tkinter import *
myk=Tk()
myk.geometry('300x300')
myk.configure(bg='yellow')
myk.title("Listbox widgets in TKinter")
mylst=Listbox(myk)
mylst.grid(row=0)
mylst.insert(0,'Delhi')
mylst.insert(1,'Bombay')
mylst.insert(2,'Jalandar')
mylst.insert(3,'Pune')
mylst.insert(END,'Goa')
Label(myk,text="Enter index to delete").grid(row=0,column=1)
e1=Entry(myk,width=5)
e1.grid(row=0,column=2)
def mydel():
    mylst.delete(int(e1.get()))
Button(myk,text='DELETE',command=mydel).grid(row=1,column=1)
mainloop()

```

```python
# Menu Button
from tkinter import *
from tkinter.filedialog import askopenfilename

myw = Tk()
myw.geometry('300x200')
myw.configure(bg='yellow')
myw.title("Menu widget using tkinter")
mymenu = Menu(myw)
myw.config(menu=mymenu)
filemenu = Menu(mymenu)
mymenu.add_cascade(label='File', menu=filemenu)
filemenu.add_command(label='New')
filemenu.add_command(label='Open')
filemenu.add_separator()
filemenu.add_command(label='Save')
filemenu.add_command(label='SaveAs')
editmenu = Menu(mymenu)
mymenu.add_cascade(label='Edit', menu=editmenu)
editmenu.add_command(label='Cut')
editmenu.add_command(label='Copy')
editmenu.add_command(label='Paste')

myw.mainloop()
```

```python
# radio button
from tkinter import *
myw = Tk()
myw.geometry('300x200')
myw.configure(bg='yellow')
myw.title("Radiobutton widget using tkinter")
myvar = IntVar()
myl = Label(myw, width=18)
myl.grid(row=0)
def myf():
    myl.configure(text="You have selected " + str(myvar.get()))

rb1 = Radiobutton(myw, text="Male", variable=myvar, value=1, width=15, anchor='w', command=myf)
rb1.grid(row=1)

rb2 = Radiobutton(myw, text="Female", variable=myvar, value=2, width=15, anchor='w', command=myf)
rb2.grid(row=2)
# rb2.select()

mainloop()
```
