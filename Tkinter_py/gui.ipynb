```python
import tkinter as tk
from tkinter import END, IntVar, Toplevel, messagebox 
from tkinter import PhotoImage, Label, Button


root=tk.Tk()
root.title("GUI")
root.configure(bg="lightblue")
root.geometry("500x500")

men=tk.Menu(root)
root.config(menu=men)
file_menu=tk.Menu(men, tearoff=0)
men.add_cascade(label="file",menu=file_menu)

file_menu.add_command(label="open")
file_menu.add_separator()
file_menu.add_command(label="new")
file_menu.add_separator()

file_menu.add_command(label="save")
file_menu.add_separator()
file_menu.add_command(label="save as")
file_menu.add_separator()
file_menu.add_command(label="edit")
file_menu.add_separator()
file_menu.add_command(label="exit")


edit_menu=tk.Menu(men, tearoff=0)

men.add_cascade(label="edit",menu=edit_menu)

edit_menu.add_command(label="open")
edit_menu.add_separator()
edit_menu.add_command(label="new")
edit_menu.add_separator()
edit_menu.add_command(label="save")
edit_menu.add_separator()
edit_menu.add_command(label="save as")
edit_menu.add_separator()
edit_menu.add_command(label="edit")
edit_menu.add_separator()
edit_menu.add_command(label="exit")

lb1=tk.Label(root,text="user",font=("Arial",8),bg="yellow",fg="red",padx=110,pady=10,relief="raised")

en1=tk.Entry(root)
en1.grid(row=0,column=1,columnspan=1)
lb1.grid(row=0,column=0,columnspan=1)
lb2=tk.Label(root,text="password",font=("Arial",8),bg="yellow",fg="red",padx=102,pady=10,relief="raised")
lb2.grid(row=1,column=0,columnspan=1)
en2=tk.Entry(root)
en2.grid(row=1,column=1,columnspan=1)
var=IntVar()
choice=tk.Label(root)
lst=tk.Listbox(root)
lst.grid(row=5)
lst.insert(0,"employed")
lst.insert(1,"Student")
lst.insert(2,"Student - employed")
lst.insert(3,"startup")
lst.insert(END,"unemployed")

lslb=tk.Label(root)

enlst=tk.Entry(root)
enlst.grid(row=6,column=1)
lslb.grid(row=6,column=0,columnspan=1)
def lst_fn():
     # messagebox.showinfo("t",enlst.get())
     lslb.configure(text=enlst.get())
lstbtn=tk.Button(root,text="list button",command=lst_fn).grid(row=7)


def show():
     txt = "Male" if var.get() == 1 else "Female"
     choice.configure(text="you have selected "+txt,bg="white",fg="blue")
lb3=tk.Label(root,text="gender -- " ,font=("Arial",8),bg="yellow",fg="red",padx=110,pady=10,relief="raised")
rb1=tk.Radiobutton(root,variable=var,command=show,value=1,text="male")
rb2=tk.Radiobutton(root,variable=var,command=show,value=2,text="female")
lb3.grid(row=2,column=0)
rb1.grid(row=2,column=1)
rb2.grid(row=2,column=2)
choice.grid(row=3,columnspan=20)

def hi():
    messagebox.showerror("title","message")
btn=tk.Button(root,text="Submit",command=hi)
btn.grid(row=10,column=0,columnspan=2)


box = Toplevel()
box.title("Image Viewer")
image_path = "img.jpg"
img = PhotoImage(file=image_path)
label = Label(box, image=img)
label.pack()
close_button = Button(box, text="Close", command=box.destroy)
close_button.pack()
box.mainloop()


# messagebox.showerror(title="title",message="message")
root.mainloop()
```
